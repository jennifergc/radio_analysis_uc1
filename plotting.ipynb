```python
import aplpy
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.wcs import WCS

# Archivos FITS
fits_file = "/home/jennifer/casa/M17_P2019/ALMA01480118_m0_nomask.fits"
contour_file = "/home/jennifer/casa/M17_P2019/ALMA01480118_m1_3sigma.fits"

# Extraer solo la capa 2D de la imagen base y corregir el WCS
with fits.open(fits_file) as hdul:
    data_base = hdul[0].data.squeeze()  # Convertir a 2D
    header_base = hdul[0].header
    wcs_base = WCS(header_base).celestial  # Extraer solo el WCS espacial

# Guardar la imagen base con WCS corregido
fits.writeto("imagen_base_2d.fits", data_base, wcs_base.to_header(), overwrite=True)

# Extraer solo la capa 2D de la imagen de contornos y corregir el WCS
with fits.open(contour_file) as hdul:
    data_contour = hdul[0].data.squeeze()  # Convertir a 2D
    header_contour = hdul[0].header
    wcs_contour = WCS(header_contour).celestial  # Extraer solo el WCS espacial

# Guardar la imagen de contornos con WCS corregido
fits.writeto("imagen_contornos_2d.fits", data_contour, wcs_contour.to_header(), overwrite=True)

# Crear la figura con APLpy usando las imágenes corregidas
fig = plt.figure(figsize=(8, 6))
f = aplpy.FITSFigure("imagen_base_2d.fits", figure=fig)

# Mostrar la imagen con colores invertidos (opcional)
f.show_colorscale(cmap='inferno', stretch='linear')

# Agregar contornos desde la imagen de contornos
f.show_contour("imagen_contornos_2d.fits", levels=[-1,1,3,5, 15, 50,100,200], colors='white', linewidths=1.2)

# Agregar etiquetas y grid
f.axis_labels.set_xtext('Ascensión Recta (RA)')
f.axis_labels.set_ytext('Declinación (Dec)')
f.add_grid()

# Mostrar la imagen
plt.show()
```

```python
import aplpy
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from astropy.wcs import WCS

# Archivos FITS
fits_file = "/home/jennifer/casa/M17_P2019/ALMA01480118_m0_nomask.fits"
contour_file = "/home/jennifer/casa/M17_P2019/ALMA01480118_m1_3sigma.fits"

# Extraer solo la capa 2D de la imagen base y corregir el WCS
with fits.open(fits_file) as hdul:
    data_base = hdul[0].data.squeeze()
    header_base = hdul[0].header
    wcs_base = WCS(header_base).celestial

fits.writeto("imagen_base_2d.fits", data_base, wcs_base.to_header(), overwrite=True)

# Extraer solo la capa 2D de la imagen de contornos y corregir el WCS
with fits.open(contour_file) as hdul:
    data_contour = hdul[0].data.squeeze()
    header_contour = hdul[0].header
    wcs_contour = WCS(header_contour).celestial

fits.writeto("imagen_contornos_2d.fits", data_contour, wcs_contour.to_header(), overwrite=True)

# 📌 Calcular sigma de la imagen de contornos (descartando NaNs)
sigma = np.nanstd(data_contour)  # Desviación estándar como estimación de sigma
sigma = 8e-3  # Desviación estándar como estimación de sigma


# 📌 Definir niveles de contorno como múltiplos de sigma
contour_levels = sigma * np.array([3, 5, 10, 20, 40, 80])  # Ajusta los valores si es necesario

print(f"📌 Sigma estimado: {sigma:.5f}")
print(f"📌 Niveles de contorno: {contour_levels}")

# Crear la figura con APLpy usando las imágenes corregidas
fig = plt.figure(figsize=(8, 6))
f = aplpy.FITSFigure("imagen_base_2d.fits", figure=fig)

# Mostrar la imagen con colores invertidos (opcional)
f.show_colorscale(cmap='inferno', stretch='linear')

# Agregar contornos desde la imagen de contornos
f.show_contour("imagen_contornos_2d.fits", levels=contour_levels, colors='white', linewidths=1.2)

# Agregar etiquetas y grid
f.axis_labels.set_xtext('Ascensión Recta (RA)')
f.axis_labels.set_ytext('Declinación (Dec)')
f.add_grid()

# Mostrar la imagen
plt.show()
```
