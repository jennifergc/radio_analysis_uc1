```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# Cargar el archivo FITS
fits_file = "ALMA01480118_contours.fits"  # Aseg√∫rate de que el nombre es correcto
hdu = fits.open(fits_file)
data = hdu[0].data  # Extraer datos de la imagen

# Verificar las dimensiones
print("Dimensiones originales del FITS:", data.shape)

# Si la imagen tiene m√°s de 2 dimensiones, reduce la matriz
data = np.squeeze(data)  # Elimina dimensiones de tama√±o 1

print("Dimensiones despu√©s de squeeze:", data.shape)

# Verifica que ahora tenga solo 2 dimensiones
if data.ndim != 2:
    raise ValueError(f"La imagen sigue teniendo {data.ndim} dimensiones, verifica el archivo FITS.")

# Graficar la imagen con contornos
plt.figure(figsize=(8, 6))
#plt.imshow(data, origin='lower', cmap='gray')  # Muestra la imagen en escala de grises
#plt.colorbar(label="Intensidad")  # Barra de color
plt.contour(data, levels=10, colors='red')  # Agrega contornos en rojo
plt.xlabel("Pixels X")
plt.ylabel("Pixels Y")
plt.title("Contornos de la imagen FITS")
plt.show()

# Cerrar el archivo FITS
hdu.close()

```
